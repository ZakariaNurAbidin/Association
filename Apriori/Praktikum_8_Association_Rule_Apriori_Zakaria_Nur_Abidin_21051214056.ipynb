# Memuat dataset

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#Memuat dataset
df = pd.read_excel('belanja_v2.xlsx')
df
```

```python
df.shape
```

```python
df.columns
```

# Pembuatan model

**minimium support=0,01 dan minimum confidence =0.3**

```python
min_support_value = 0.01
frq_items = apriori(df, min_support=min_support_value, use_colnames=True)

min_confidence_value = 0.3
rules = association_rules(frq_items, metric="confidence", min_threshold=min_confidence_value)
rules = rules.sort_values(['confidence', 'lift'], ascending=[False, False])

# Menampilkan hasil aturan asosiasi
print(rules)
```
